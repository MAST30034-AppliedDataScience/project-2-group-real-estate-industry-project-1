```python
df_live = pd.read_csv('../../data/curated/Liveability_score/Liveability_by_suburb_overall.csv')
df_afford = pd.read_csv('../../data/curated/affordable_score_by_suburb.csv')
```

```python
# Merge the two dataframes on 'Suburb'
df_combined = pd.merge(df_live, df_afford, on='Suburb')

# Define the weights for liveability and affordability scores
live_weight = 0.4  # 你可以根据需求修改
afford_weight = 0.6  # 你可以根据需求修改

# Calculate the live_afford_score
df_combined['live_afford_score'] = (df_combined['Liveability score'] * live_weight) + (df_combined['Final Affordable Score'] * afford_weight)

# Sort the dataframe by 'live_afford_score' in descending order
df_sorted = df_combined.sort_values(by='live_afford_score', ascending=False)

# Display the sorted dataframe
df_sorted.head(20)
```
