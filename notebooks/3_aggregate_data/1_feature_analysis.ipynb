```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
suburb_aggregates_df = pd.read_csv("../../data/curated/aggregate_property_suburb.csv")
```

```python


# Drop unnecessary columns like 'Unnamed: 0' and 'Suburb' for correlation analysis
numeric_data = suburb_aggregates_df.drop(columns=['Unnamed: 0', 'Suburb'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap to visualize the correlation, focusing on 'avg_Cost' as the response variable
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0, linewidths=1, linecolor='black')

# Highlight the correlation with 'avg_Cost'
plt.title('Correlation Heatmap with avg_Cost as Response', fontsize=16)
plt.show()
```
